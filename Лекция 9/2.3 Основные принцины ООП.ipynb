## 1. Объектно-ориентированное программирование как парадигма

### Парадигмы программирования

В программировании существуют несколько основных подходов, которые определяют, как программисты структурируют и организуют свой код. Они называются парадигмами. **Парадигма** — набор правил и критериев, которые помогает стандартизировать написание кода, что снижает риск ошибок, ускоряет разработку и делает код более понятным для других программистов.

Самый простой и прямолинейный подход — это **процедурное** программирование. Программы в этой парадигме организованы как цепочки команд, выполняемых последовательно. Это что-то вроде кулинарного рецепта, который надо выполнить по порядку для получения желаемого блюда. Есть более общий вариант, **структурное** программирование — методология, которая позволяет писать программу как набор блоков, которыми являются ветвления, циклы, последовательность операторов и процедуры. Ему мы и придерживались в течение этого курса.

Также есть и **функциональное** программирование. Этот подход основан на концепции математических функций, которые принимают входные данные и возвращают результат. Существует и ещё несколько других парадигм.

Однако сейчас одним из наиболее популярных подходов является **объектно-ориентированное программирование** (ООП), где код организуется вокруг объектов и классов.




### Особенности ООП

Основная идея объектно-ориентированного программирования заключается в том, чтобы представить программу в виде совокупности взаимодействующих объектов, каждый из которых является экземпляром определённого класса.

С первого раза звучит непонятно. Разберем такой пример: нужно написать для интернет-магазина каталог товаров. Если руководствоваться принципами ООП, в первую очередь нужно создать объекты: карточки для товаров. Потом заполнить эти карточки данными: названием товара, свойствами, ценой. И потом прописать доступные действия: изменение объекта или взаимодействие с ним.

Если использовать функциональную парадигму, то чтобы сделать каталог товаров, начинать нужно не с карточек, а с функций, заполняющих эти карточки. То есть объект будет не отправной точкой, а будет результатом работы функции. Обычно написать функцию быстрее, чем создавать объекты и прописывать взаимодействие между ними. Но если объём кода большой, работать с разрозненными функциями будет довольно сложно, поэтому чем больше проект — тем выгоднее в нём использование парадигмы ООП.

В коде, написанном по принципам ООП, выделяют четыре основных элемента:
1. **Объект**.  
Это часть кода, которая описывает элемент с конкретными характеристиками и функциями. Карточка товара в каталоге интернет-магазина — это объект, и кнопка «заказать» — тоже.
2. **Класс**.  
Это шаблон, на базе которого можно построить объект в программировании. Например, у интернет-магазина может быть класс «Карточка товара», который описывает общую структуру всех карточек. И уже из него создаются конкретные карточки — объекты.  
Классы могут наследоваться друг от друга. Например, есть общий класс «Карточка товара» и вложенные классы, или подклассы: «Карточка телевизора», «Карточка ноутбука», «Карточка телефона». Подкласс берёт свойства из родительского класса, например, цену товара, количество штук на складе или производителя. При этом имеет свои свойства, например, диагональ дисплея для «Карточки телевизора» или количество сим-карт для «Карточки телефона».
3. **Метод**.  
Это функция внутри объекта или класса, которая позволяет взаимодействовать с ним или другой частью кода. В примере с карточками товара метод может:  
● Заполнять карточку конкретного объекта нужной информацией.  
● Обновлять количество товара в наличии, сверяясь с базой данных.  
● Удалять товар.  
● Сравнивать и предлагать купить похожие товары.
4. **Атрибут**.  
Это характеристики объекта в программировании — например, цена, производитель или объём оперативной памяти. В классе прописывают какие атрибуты есть, а в объектах с помощью методов заполняют эти атрибуты данными. Всё просто.

### Реализация ООП в языке Python

Теперь посмотрим, как эта абстрация ложится на реальный язык программирования, в нашем случае — Python.

Итак, класс в программировании — это шаблон или чертёж для создания объектов (экземпляров класса). Класс описывает свойства (атрибуты) и действия (методы), которыми могут обладать все объекты этого класса.

Для описания класса в Python используется ключевое слово `class`, имя класса, двоеточие и тело класса с отступом.

Простейший класс в Python может выглядеть так:
```
class ClassName:  # Ключевое слово class и имя класса.
    # Тело класса
    ...
```
**Важно**. Согласно стандарту PEP8 названия классов дают в единственном числе и пишут с большой буквы, без подчеркивания.

В теле класса обычно объявляют атрибуты, конструктор `__init__()` и другие методы. Конструктор класса `__init__()` — это специальный метод, который автоматически вызывается при создании нового объекта класса. Методы, начинающиеся с двойного подчеркивания, также называются магическими методами, поскольку они вызываются сами при определенных условиях.

Давайте создадим класс, представляющий простую библиотечную систему для управления книгами. И сразу напишем конструктор, создающий новую книгу.

````python
class Book:
    def __init__(self, title, author):
        self.title = title         # Название
        self.author = author       # Автор
        self.is_available = True   # доступна ли книга в библиотеке
````

В Python ключевое слово `self` используется внутри методов класса для ссылки на конкретный экземпляр этого класса. Оно позволяет методам класса получать доступ к атрибутам и другим методам того же экземпляра. Хотя в Python использование именно слова `self` не обязательно, это соглашение является общепринятым и очень важно для ясности и читаемости кода.

Теперь, когда у нас есть класс (то есть схема, по которой можно создавать объекты), мы можем создать два экземпляра этого класса:

````python
book1 = Book("Чистый код", "Р. Мартин")
book2 = Book("Война и мир", "Л. Толстой")

print(book1)
print(book2)
````

Здесь мы вызвали класс точно так же, как вызываем функции, и передали в качестве аргументов им название книги и автора. Обратине внимание, аргумент `self` мы при этом проигнорировали. В результате было создано два разных объекта типа Book.

Теперь попробуем расширить наш класс и добавить удобное строковое описание каждой книги с помощью специального метода `__str__`:

````python
class Book:
    def __init__(self, title, author):
        self.title = title         # Название
        self.author = author       # Автор
        self.is_available = True

    def __str__(self):
        status = 'Доступна' if self.is_available else 'Выдана'
        return f"'{self.title}', {self.author} - {status}"

book1 = Book("Чистый код", "Р. Мартин")
book2 = Book("Война и мир", "Л. Толстой")

print(book1)
print(book2)
````

Видно, что в переменную `self.title` была положена строка с названием книги из соответствующего экземпляра класса.  
Давайте посмотрим что будет, если в методе `__str__` мы уберем обращение через self:

````python
class Book:
    def __init__(self, title, author):
        self.title = title         # Название
        self.author = author       # Автор
        self.is_available = True

    def __str__(self):
        status = 'Доступна' if is_available else 'Выдана'
        return f"'{title}', {author} - {status}"

book1 = Book("Чистый код", "Р. Мартин")
print(book1)
````

Мы получили ошибку: ведь компилятор посчитал `is_available` обычной глобальной переменной, а не атрибутом нашего класса, и такой переменной не существует.

Теперь добавим в наш класс несколько наших собственных (не зарезервированных) методов. Сделаем методы, которые меняют статус книги когда её кто-то берет, и когда кто-то возвращает. Методы именуются через подчеркивание, как обычные функции. А также добавим общий для всех экземпляров атрибут `item_type`:

````python
class Book:
    item_type = "Книга"

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        status = 'Доступна' if self.is_available else 'Выдана'
        return f"{self.item_type} '{self.title}', {self.author} - {status}"

    def check_out(self):
        if self.is_available:
            self.is_available = False
            return True
        else:
            return False

    def return_book(self):
        if not self.is_available:
            self.is_available = True
            return True
        else:
            return False
````

Попробуем взять одну из книг:

````python
book1 = Book("Чистый код", "Р. Мартин")
print(book1)

book1.check_out()  # вот так вызывается метод класса в явном виде через его экземпляр
print(book1)

book1.return_book()  # а теперь вернем книгу
print(book1)
````

По умолчанию все методы в классе привязаны к экземпляру класса, а не к самому классу. Поэтому методы, которые мы добавили, называются **методами экземпляра**. Однако с помощью декоратора `@classmethod` можно объявить метод, который привязан к самому классу, а не к его экземпляру. Такие методы называют **методами класса**. Аналогично аргументу `self`, здесь общепринято использовать аргумент `cls`.

**Важно**. В Python декораторы — это способ изменить или расширить поведение функций или классов без непосредственного изменения их исходного кода. Они представляют собой ключевое слово, которое помещаются непосредственно в строку перед объявлением функции.

Помимо методов класса и экземпляра можно объявлять **статические методы**. Они описываются при помощи декоратора `@staticmethod`. Такие методы в основном используются как вспомогательные функции и работают с данными, которые им переданы.

Попробуем добавить всё это в наш класс:

````python
class Book:
    item_type = "Книга"  # общий атрибут

    # Конструктор:
    def __init__(self, title, author):
        # атрибуты экземпляра
        self.title = title
        self.author = author
        self.is_available = True

    # Метод экземпляра:
    def __str__(self):
        status = 'Доступна' if self.is_available else 'Выдана'
        return f"{self.item_type} '{self.title}', {self.author} - {status}"

    # Метод экземпляра:
    def check_out(self):
        if self.is_available:
            self.is_available = False
            return True
        else:
            return False

    # Метод экземпляра:
    def return_book(self):
        if not self.is_available:
            self.is_available = True
            return True
        else:
            return False

    # Метод класса:
    @classmethod
    def get_item_type(cls):
        return cls.item_type

    # Статический метод:
    @staticmethod
    def get_greeting():
        return 'Добро пожаловать в библиотеку!'
````

Такие методы можно вызывать непосредственно из класса, даже не пользуясь конкретным экземпляром:

````python
print(Book.get_item_type())  # Метод класса
print(Book.get_greeting())   # Статический метод
````

Теперь, добавляя новые атрибуты и методы, мы можем расширить наш класс до функциональности, способной обслужить целое здание библиотеки с тысячами книг. Однако уже здесь виден основной недостаток ООП — необходимость создавать отдельные методы на каждое действие очень сильно раздувает код в размерах и он становится довольно громоздким даже для небольшого класса.

## 2. Основные принципы ООП

Объектно-ориентированное программирование в свою очередь тоже бывает разным и может строиться на разных принципах. Но самой популярной до сих пор остается концепция, основанная на этих трёх принципах: наследование, полиморфизм и инкапсуляция.

### Наследование

В этом принципе — вся суть объектно-ориентированного программирования.
Разработчик создаёт:   
● Класс с определёнными свойствами.  
● Подкласс на его основе, который берёт свойства класса и добавляет свои.  
● Объект подкласса, который также копирует его свойства и добавляет свои.

Каждый дочерний элемент наследует методы и атрибуты, прописанные в родительском. Он может использовать их все, отбросить часть или добавить новые. При этом заново прописывать эти атрибуты и методы не нужно.

Для примера создадим класс Shape, который будет **родительским** классом для фигур на плоскости. А также создадим два **дочерних** класса с конкретными фигурами:

````python
# Родительский класс для фигур
class Shape:
    color = "красный"  # общий для фигур параметр

    def __init__(self):
        pass

    def __str__(self):
        return "Плоская фигура класса Shape"

# Класс-наследник от Shape для работы с прямоугольниками
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

# Класс-наследник от Shape для работы с кругом
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
````

У нас получился родительский класс с общим для всех фигур атрибутом color и методом `__str__`, которые мы **наследуем** в наших подклассах. Однако метод `__init__` уже определяется для каждого подкласса свой.

````python
rectangle = Rectangle(10, 5)  # создаём экземпляр прямоугольника
print(rectangle)

circle = Circle(10)  # cоздаём экземпляр круга
print(circle.color)
````

Видно, что хоть мы и создавали объекты с помощью классов конкретной фигуры, мы имеет доступ и ко всем методам и атрибутам родительского класса Shape. Причем в классе Rectangle мы вправе сделать ещё один подкласс, делая множественные цепочки подклассов со сложной иерархией там где это нужно.

В этом и заключается принцип наследования. Он позволяет обойтись без дублирования и хранить общий код для всех дочерних классов в одном месте.

### Инкапсуляция

Этот принцип гласит, что вся информация, которая нужна для работы конкретного объекта, должна храниться внутри этого объекта. Если нужно вносить изменения, методы для этого тоже должны лежать в самом объекте — посторонние объекты и классы этого делать не могут. Для внешних объектов должны быть доступны только публичные атрибуты и методы.

Посмотрим на наш класс Circle. Оказывается, мы можем обратиться к атрибуту класса напрямую. И даже изменить его:

````python
circle = Circle(10)
print(circle.radius)
circle.radius = 15
print(circle.radius)
````

К сожалению, в языке Python нет способа закрыть доступ к атрибуту для редактирования извне. Однако существует соглашение, по которому метод или атрибут, имя которого начинается с подчёркивания, является защищённым. Использование подчеркивания — это норма поведения, подразумевающая, что другие программисты должны понимать и напрямую не обращаться к этим атрибутам или методам.

Давайте сделаем другим программистам сигнал о том, что атрибуты наших фигур являются приватными и добавим двойное подчеркивание к `__radius`. Его следует получать или изменять только через методы класса.

````python
# Родительский класс для фигур
class Shape:
    color = "красный"  # общий для фигур параметр

    def __init__(self):
        pass

    def get_params(self):
        pass

    def set_params(self):
        pass

    def get_color(self):
        return self.color

# Класс-наследник от Shape для работы с кругом
class Circle(Shape):
    def __init__(self, radius):
        self.__radius = radius

    # Метод для получения радиуса извне
    def get_params(self):
        return self.__radius

    # Метод для изменения радиуса извне
    def set_params(self, radius):
        # добавим простую проверку
        if radius >= 0:
            self.__radius = radius
        else:
            raise ValueError("Радиус не может быть отрицательным!")
````

Попробуем создать круг и обновить его радиус, передав отрицательное значение. Получаем соответствующую ошибку:

````python
circle = Circle(10)
print(f"Я круг радиуса {circle.get_params()}, цвета: {circle.get_color()}")
circle.set_params(-5)
````

Таким образом мы сделали радиус нашей фигуры приватным параметром, а чтобы получать или обновлять его значение добавили соответствующие публичные методы.

Принцип инкапсуляции призван обеспечить безопасность, чтобы не повредить данные внутри какого-то класса со стороны. Ещё он помогает избежать случайных зависимостей, когда из-за изменения одного объекта что-то ломается в другом.

### Полиморфизм

Согласно этому принципу, один и тот же метод может работать по-разному в зависимости от объекта и данных, которые ему передали. Этот принцип так же, как и другие, обеспечивает отсутствие ошибок при использовании объектов.

Давайте создадим метод, который будет по-разному вычислять площадь в зависимости от того, для какой фигуры его вызвали:

````python
import math

# Родительский класс для фигур
class Shape:
    def __init__(self):
        pass

    def calculate_area(self):
        pass

# Класс-наследник от Shape для работы с прямоугольниками
class Rectangle(Shape):
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    def calculate_area(self):
        return self.__length * self.__width

# Класс-наследник от Shape для работы с кругом
class Circle(Shape):
    def __init__(self, radius):
        self.__radius = radius

    def calculate_area(self):
        return math.pi * self.__radius ** 2
````

Тестируем:

````python
rectangle = Rectangle(10, 5)  # создаём экземпляр прямоугольника
rectangle_area = rectangle.calculate_area()
print(f"Площадь прямоугольника: {rectangle_area}")

circle = Circle(10)  # cоздаём экземпляр круга
circle_area = circle.calculate_area()
print(f"Площадь круга: {circle_area}")
````

Принцип полиморфизма позволяет использовать один и тот же интерфейс для различных типов объектов. В нашем примере метод `calculate_area` можно вызывать для вычисления площади разных типов фигур.

Ещё один популярный пример полиморфизма — операция сложения "+", которая работает по-разному для целых чисел и для строк. Если мы попытаемся сложить два числа, неявно вызывается магический метод `__add__`, отвечающий за сложение объектов типа int и возвращающий результат арифметической операции. А если попытаемся сложить две строки, то будет вызван уже метод класса str, возвращающий в качестве результата конкатенацию строк:

## 3. Преимущества и недостатки ООП

Настало время поговорить о преимуществах и недостатках ООП.

Преимущества:

● Код проще читать.  
Даже в чужом коде обычно сразу видны конкретные объекты и методы, их удобно искать и смотреть, что именно они делают.  
● Код легче обновлять.  
Класс или метод достаточно изменить в одном месте, чтобы он изменился во всех наследуемых классах и объектах. Не нужно переписывать каждый объект отдельно, выискивая, где именно в коде он расположен.  
● Код можно переиспользовать.  
Один раз написанный класс или объект можно затем переносить в другие проекты.   
● Удобнее работать в команде.  
Разные люди могут отвечать за разные объекты, не мешая друг другу и при этом пользуясь трудами коллег.  
● Наличие шаблонов проектирования.  
Именно на базе ООП построены готовые решения для взаимодействия классов друг с другом, которые позволяют не писать этот код с нуля, а взять готовый шаблон.

Недостатки:

● Сложность в освоении.  
ООП сложнее, чем функциональное программирование. Для написания кода в этой парадигме нужно знать гораздо больше, поэтому перед созданием первой рабочей программы придётся освоить много информации: разобраться в классах и наследовании, научиться писать публичные и внутренние функции, изучить способы взаимодействия объектов между собой.  
● Громоздкость кода.  
Там, где в функциональном программировании хватит одной функции, в ООП нужно создать класс, объект, методы и атрибуты. Для больших программ это плюс, так как структура будет понятной, а для маленьких может оказаться лишней тратой времени.  
● Низкая производительность.  
Объекты потребляют больше памяти, чем простые функции и переменные. Скорость компиляции и интерпретации тоже падает, иногда очень сильно! Поэтому перед обработкой очень большого массива данных для анализа следует задуматься, стоит ли реализовывать её через методы ООП.

В итоге мы получаем, что основная функция объектно-ориентированного программирования — это облегчить написание больших и сложных программ, над которыми трудятся группы разработчиков.

Поэтому данная тема включена в рамки курса по анализу данных не столько для того, чтобы заставить переходить на принципы ООП при написании своих программ, сколько для понимания внутренней структуры больших библиотек для анализа и принципов работы с ними. А также эти знания могут пригодится в будущем для работы в большой команде с кодом от других людей.
