# 2. Установка интерпретатора Python

## Основные понятия

**Язык программирования** — это набор формальных правил, по которым компьютер может понимать команды (инструкции) и выполнять их. Языки программирования придуманы искусственно, поэтому в них строго определены  порядок следования элементов (синтаксис) и смысл этих элементов (семантика). **Прогрaммa** — это последовaтельность инструкций, которые компьютер выполняет во время работы. Текст программы на любом языке программирования называется программным **кодом**.


Языки программирования бывают **компилируемые** и **интерпретируемые**. Если программа написана на компилируемом языке (**C**, **C++**, **Pascal**), то перед выполнением её нужно полностью проверить на наличие синтаксических ошибок и уже после этого перевести в понятную для компьютера форму — машинный код. Это делает специальная программа, которая называется компилятором. Результатом компиляции является исполняемый файл, который может быть запущен без наличия исходного кода.

Если программа написана на интерпретируемом языке (**Python**, **PHP**, **Ruby**), она не переводится в машинный код целиком. Вместо этого специальная программа, которая называется **интерпретатором**, проходит по коду, анализирует его и выполняет каждую отдельную строку или команду. Исходный код интерпретируемых языков более переносим, поскольку он может быть выполнен на разных операционных системах без изменений.

Существуют языки программирования, которые совмещают оба подхода (**C#**, **Java**). В таких языках код исходной программы сначала компилируется в промежуточный код (байт-код), а уже потом, во время выполнения, переводится в машинный код.

В нашем курсе мы будем использовать язык **Python**. В настоящее время в русском языке для обозначения используют два варианта: «Питон» и «Пайтон». Первый является прямым переводом, второй (более правильный) — адаптацией английского произношения. Язык Python разработал голландский программист **Гвидо Ван Россум** (Guido van Rossum) в 1991 году. Гвидо был фанатом британского комедийного сериала «Monty Python’s Flying Circus», откуда и пришло название языка.  Автор создавал язык с ориентацией на повышение производительности разработчика, читаемости кода и его качества, а также на обеспечение переносимости между платформами написанных на нём программ. К недостатками языка можно отнести зачастую более низкую скорость работы и более высокое потребление памяти написанных на нём программ, по сравнению с аналогичным кодом, написанным на компилируемых языках.



Python относится к интерпретируемым языкам программирования, поэтому чтобы запустить написанную на Python программу, необходим интерпретатор Python.

## Команда print()

Запустим ноутбук JupyterLab или GoogleColab и напишем первую программу. Чтобы добавить текстовый блок используется кнопка **+Текст** вверху экрана. Блок поддерживает язык текстовой разметки ``Markdown`` (https://ru.wikipedia.org/wiki/Markdown).

Для вывода результатов работы команд на экран используется функция **print()**. Если вы работаете в какой-либо среде разработки, то "вывод на экран" подразумевает печать данных в *консоль* (или же, по-другому, в *терминал*).

Внутри круглых скобок команды через запятую указывается то, что необходимо вывести на экран. Если это текст, обрамляем его кавычками - одинарными, или двойными. Главное, чтобы текст начинался и заканчивался кавычками одного типа. Команда print записывается только строчными буквами, другое написание недопустимо, так как в Python строчные и заглавные буквы различны. Так же не следует использовать пробелы перед открывающейся скобкой, так как это противоречит стандарту PEP8. Стандарт <a href = "https://peps.python.org/pep-0008/"><b>PEP8</b></a> (Python Enhancement Proposal) содержит перечень принципов написания красивого и лаконичного программного кода на языке Python и служит для улучшения такого параметра как "читаемость кода". Вот <a href = "https://pythonworld.ru/osnovy/pep-8-rukovodstvo-po-napisaniyu-koda-na-python.html">ссылка</a> на переведенную на русский документацию по PEP8.

Напишем простейшую программу. На экран будет выведена строка "Hello, World!". Добавим в ноутбук следующий код с помощью кнопки **+Код**:

````python
print('Hello, World!')
````

Ячейка с кодом запускается при помощи стрелочки **>** рядом с ячейкой, либо через сочетание клавиш **Ctrl + Enter**.

Немного изменим программу:

````python
print('Hello,', 'World!')
````

В данном случае мы внутри команды указали две строки, каждую заключили в кавычки и разделили запятой. Таким образом, функция может принимать для печати неограниченное количество агрументов. **Внимание:** после запятой согласно стандарту PEP8 обязательно нужно добавлять пробел.

В ячейке с кодом можно запускать и программы с внешней командной оболочки (bash или powershell). Для этого используется восклицательный знак перед командой. Например, посмотрим установленную на виртуальной машине версию Python:

## Именованные аргументы print()

Рассмотрим пример:

````python
print('При')
print('вет', '!')
````

Видно, что функция print при выводе разделяет аргументы пробелами, а в конце переходит на новую строчку. Но это не всегда удобно.
Для тонкой настройки вывода у функции print существуют необязательные именованные аргументы **sep=**, **end=** .

````python
# С помощью команды help можно посмотреть информацию об интересующей функции
help(print)
````

Таким образом, значения именованных аргументов функции print по умолчанию такие: `print(..., sep=' ', end='\n').` Позже, в других уроках, мы подробнее рассмотрим как работают аргументы функций — а пока просто посмотрите примеры ниже.

**sep** — разделитель аргументов (по умолчанию пробел) и **end** — то, что выводится после вывода всех аргументов (по умолчанию символ начала новой строки). Например, если end сделать пустой строкой, то print не перейдет на новую строчку, и следующий print продолжит вывод прямо на этой же строке.

````python
print('При', end='')
print('вет')
````

````python
print('один', 'два', 'три', sep='-')
````

**Внимание.** Знак = здесь не выполняет никакого присваивания, переменных end и sep не появляется.
Согласно стандарту PEP8 пробелы вокруг знака "=" в данном случае не используются.

## Экранирующая последовательность

Если внутри кавычек встречается символ \ — обратная косая черта, обратный слеш, бэкслеш, то он вместе с идущим после него символом образует экранирующую последовательность (escape sequence) и воспринимается интерпретатором как единый специальный символ.

В частности, \n — символ начала новой строки. Кроме того, \t — табуляция, \\'  — кавычка, \\ — просто бэкслеш. Использование экранирующих последовательностей вместо специальных символов называется их экранированием.

````python
print('Замена символов на подстановки носит название "экранирование".')
print('Замена символов на подстановки носит название \'экранирование\'.')
````

Экранирование очень важно при работе со строками, если не применить его вовремя можно получить ошибку:

````python
print('Замена символов на подстановки носит название 'экранирование'.')
````

Ещё примеры:

````python
print('Восход\t04:04\nЗакат\t20:53')
````

````python
print('print(\'Восход\\t04:04\\nЗакат\\t20:53\')')
````

## Необработанные строки (raw)

Необработанная строка в Python создается путем добавления к строковому литералу префикса «r» или «R». Необработанная строка Python рассматривает обратную косую черту (\\), как буквальный символ.

````python
print("Файл на диске С:\\\\")
print(r"Файл на диске С:\\")
````

Если открывать и закрывать строку не одной, а тремя кавычками подряд, внутри можно делать обычные переводы строки (внутри одинарных кавычек так делать нельзя).

````python
print("""Компьютерное программирование
для всех.

Гвидо Ван Россум
""")
````

## Комментарии в тексте программы

Комментарии – это способ объяснить ваш код в той строке, где он написан. Возьмем следующий пример кода, в нём “выводим на экран границы светового дня” – это комментарий:

````python
print('Восход\t04:04\nЗакат\t20:53')  # выводим на экран границы светового дня
````

В языке Python комментарии начинаются с символа **#** и заканчиваются последним символом в текущей строке. При этом комментарий может занимать и всю строку целиком. По стандарту PEP8 рекомендуется менее 79 символов на одну строку комментария и хотя бы два пробела между кодом программы и символом хэша #.

Помните, что комментарии должны быть **осмысленными** и объяснять неочевидный код, а не просто дублировать написанное другими словами.

Если необходимо сделать многострочный комментарий, реализовать это можно двумя способами. Первый — просто использовать символ # на каждой строке:

````python
# МояЛучшаяПрограмма
# Версия 1.6.3
# Параметры:
#
# -t (--text): показать текстовый интерфейс
# -h (--help): показать окно помощи
````

Второй способ — использовать уже рассмотренную выше многострочную raw-строку:

````python
"""
МояЛучшаяПрограмма
Версия 1.6.3
Параметры:

-t (--text): показать текстовый интерфейс
-h (--help): показать окно помощи
"""
````

Многострочные комментарии довольно часто используют в начале Python-файла. В них указывается информация о проекте, назначении файла, параметрах, программисте, который его разработал, и другие данные.

Комментирование кода помогает объяснить ваш мыслительный процесс, а также помогает вам и другим людям понять смысл вашей программы. Ведь даже самому автору программы бывает тяжело вспомнить, как он реализовывал алгоритм, если приходится возвращаться к какому-то коду спустя долгое время. Ещё комментарии позволяют вам легче находить ошибки и исправлять их, улучшая этим код, а также повторно использовать его в других программах.


## Команда input()

Ранее наши программы выводили на экран только фразы, известные в момент написания кода. Но программы могут работать и с данными, которые станут известны только во время выполнения: например, их будет вводить пользователь с клавиатуры. Реализовать это можно так:

````python
print('Как тебя зовут?')
name = input()
print('Привет,', name)
````

Команда **input()** всегда записывается c круглыми скобками. Работает это так: когда программа доходит до места, где есть input(), она ждет, пока пользователь введет строку с клавиатуры (ввод завершается нажатием клавиши Enter). Введенная строка подставляется на место input(). Т.е. input() получает от пользователя какие-то данные и в место вызова подставляет строковое значение, в нашем случае записывает его в качестве значения переменной <b>name</b>.

<b>Примечание.</b> Текст, который мы печатаем перед считыванием данных в некоторых случаях нужен чтобы пользователь, который будет вводить данные понимал, что именно от него требуется. Поскольку это достаточно распространённая практика, то в языке Python можно выводить текст, передавая его в качестве параметра в команду input(). Предыдущий код можно переписать так:

````python
name = input('Как тебя зовут? ')
print('Привет,', name)
````
