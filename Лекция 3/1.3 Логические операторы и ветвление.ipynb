# 1. Логические выражения и арифметические операции

В естественном языке (например, русском), чтобы сравнивать одно с другим, мы используем слова "равно", "больше", "меньше". В языках программирования для этого есть специальные знаки, подобные тем, которые используются в математике.
Поэтому, чтобы составлять логические условия, в Python существует несколько стандартных логических операторов.

#### Оператор "равно"

````python
10 == 10
````

````python
10 == 10.0
````

**Внимание.** Помните, что строковые данные нельзя сравнивать с числовыми напрямую без конвертации типов:

````python
10 == "10"
````

**Ещё раз внимание.** Не путайте операцию сравнения **==** и операцию присваивания **=** . Это разные операции!

#### Оператор "не равно"

````python
2 != 3
````

#### Операторы сравнения:

````python
print(2 > 5)
print(3 <= 3)
print(7 >= 7)
print(8 < 3)
````

Позволительны и более сложные конструкции:

````python
x = 3
print(2 < x < 4)
````

#### Конвертация типов:

````python
bool(0)
````

Всё, что не ноль, является истиной:

````python
bool(-5)
````

#### Сравнение float
**Внимание**. Будьте осторожны при сравнении вещественных чисел. Из-за внутреннего устройства языка float имеют ограниченную точность. Посмотрите на следующий пример:

````python
0.1 + 1.1 == 1.2  # сравнение даёт неверный результат
````

````python
0.1 + 1.1  # при выполнении такой операции подсчет недостаточно точен
````

Поэтому при сравнении пользуйтесь округлением, например так:

````python
round(0.1 + 1.1, 1) == round(1.2, 1)
````

### Логические выражения

Логические выражения — это выражения, которые содержат в себе один или больше логических операторов.

#### Логическое "и"

````python
x, y = True, False
print(x and y)
````

#### Логическое "или"

````python
x, y = True, False
print(x or y)
````

#### Логическое отрицание

````python
y = False
print(not y)
````

#### Составные логические выражения

````python
x, y, z = True, False, True
result = x and y or z
print(result)
````

Особенность Python — так называемая "ленивая" обработка логических выражений and и or. Пример:

````python
x = 13
y = False

print(x or y)
````

В результате работы этого выражения мы видим, что на экране не True, как мы могли бы ожидать, а число 13.

Python, начиная интерпретировать логическое выражение, видит, что "x" является истинным, и понимает, что ему не нужно выполнять оставшуюся часть логического выражения. Поэтому раз "x" — это истина, затем стоит оператор or - логическое сложение, значит нас устраивает то, что мы можем остановиться в этот момент и результатом выполнения выражения будет как раз значение "x", без проверки переменной "y".

````python
x = 321
y = "boom"

print(x and y)
````

В результате работы логического выражения "x and y" мы получаем как результат работы логического выражения строку "boom". Python выполняет логическое выражение, до тех пор, пока оно имеет смысл, и результатом является последнее значение.

## 2. Условный оператор if

На предыдущих уроках мы научились писать программы, в которых выполнялись все строки по очереди. Однако очень часто необходимо, чтобы код выполнялся только при определённом условии. Такая концепция в программировании, которая позволяет выбирать, какой блок кода выполнить, в зависимости от выполнения определенных условий, называется **ветвление в коде**.

В таком случае используется условный оператор. В Python, как и во многих языках, он обозначается словом **if**. Условный оператор позволяет в буквальном смысле разделить выполняемый код на несколько веток:

изображение.png

Общий синтаксис условного оператора в языке Python следующий:
```
if <условие1 (булевское выражение)>:
    <код, который выполнится, если условие верно>
elif <условие2 (булевское выражение)>:
    <код, который выполнится, если условие1 было неверно, а условие2 верно>
else:
    <код, который выполнится, если условие1 и условие2 были неверны>
```
Для записи используются ключевые слова **if**, **else** и **elif** («если», «иначе», «иначе если»), двоеточие и  отступ в четыре пробела. Пример простого оператора if:

````python
x = 5
if x < 0:
    print("x отрицателен")

# Следующий код уже не входит в условие, потому что записан без отступа в 4 пробела, поэтому он выполнится в любом случае
print("Готово.")
````

Примечание. Jupyter Notebooks позволяет использовать отступы в два пробела для большей компактности кода. Это можно поменять в меню Инструменты -> Настройки -> Редактор.

### if - else

Оператор else позволяет выполнить какой-либо код, если условие в блоке if не выполнилось:

````python
x = 5
if x < 0:
    print("x отрицателен")
else:
    print("x положителен")
````

### if - elif - else

Оператор elif используется, когда нужно проверить несколько разных условий друг за другом.

elif — это короткая запись для «else: if». Если не пользоваться короткой записью, if пришлось бы писать на отдельной строчке и с отдельным отступом (а все, что внутри этого if, — с дополнительным отступом). Это не очень удобно, и elif избавляет от такой необходимости:

````python
x = 0
if x < 0:
    print("x отрицателен")
elif x == 0:
    print("x равен нулю")
else:
    print("x положителен")
````

Однако никто не запрещает использовать оператор if внутри оператора if. Это называется **вложенным** оператором if. Мы можем вкладывать несколько слоев операторов if, elif, else.

Обратите внимание, что код, который должен выполняться внутри каждого условия, обязательно записывается с отступом в 4 пробела от уровня if, elif и else: в питоне области видимости переменных обозначаются отступами. То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где он заканчивается.

Посмотрим что будет, если допустить ошибку в синтаксисе:

````python
x = 5

if x < 0:
    print("x отрицателен")
     print("условие выполнено")
else:
    print("x положителен")
````

Интерпретатор выдаёт ошибку "неожиданный отступ".

Условный оператор if является **фундаментальным элементом программирования** и позволяет создавать в коде разветвленные логические структуры.

### Операции сравнения и логические операции

В качестве условия может выступать результат операций сравнения:
*    \> (больше);
*    \>= (больше или равно);
*    < (меньше);
*    <= (меньше или равно);
*    == (равно);
*    != (не равно).

Для записи сложных условий можно применять логические операции:
*    and — логическое «И» для двух условий. Возвращает True, если оба условия истинны, иначе возвращает False;
*    or — логическое «ИЛИ» для двух условий. Возвращает False, если оба условия ложны, иначе возвращает True;
*    not — логическое «НЕ» для одного условия. Возвращает False для истинного условия, и наоборот.

В if можно подставлять и более сложные булевские выражения:

````python
x = 7
if (x > 3 and x < 11) or x == 15:
    print("Выполнено")
````

В логическом операторе можно использовать двойное неравенство. Пример выше лучше будет записать так:

````python
x = 7
if (3 < x < 11) or x == 15:
    print("Выполнено")
````

Помните, что для сравнения на равенство в Python используется оператор ==, а не =. Первый применяется для сравнения значений, а второй - для присваивания значений переменным. Это частая ошибка начинающих в Python.

### Приоритет операций

Коснёмся здесь приоритета выполнения операций при проверке условий. Следующая таблица определяет приоритет выполнения от высшего к низшему:

Операция | Описание
--- | ---
() | скобки
** | возведение в степень
+x, -x, ∼x | унарный плюс, унарный минус, битовое НЕ
*, /, % | умножение, деление, остаток от деления
+, - | сложение, вычитание
<<,>> | битовые сдвиги
& | битовое И
ˆ | битовое исключающее ИЛИ
\| | битовое ИЛИ
in, not in, is, is not, <, <=, >,>=,<>, !=,== | проверка на принадлежность, на идентичность, сравнения
not | логическое НЕ
and | логическое И
or | логическое ИЛИ

Как видно, скобки имеют наивысший приоритет. При этом все операции, имеющие одинаковый приоритет выполняются слева направо. Исключениями являются сравнения и операция возведения в степень, которая выполняется справа налево. Таким образом 2\*\*3\*\*2 понимается как 2\*\*(3\*\*2).

Примеры:

````python
c = 2
d = 4
result = d / c * 2
# Идём слева направо: сначала деление, потом умножение на результат
print(result)
````

````python
x = 5
y = 10
result = x < y and x * 2 == y
# Сначала умножение, затем сравнение и потом логическое И: (5 * 2 == 10) and (5 < 10) = True
print(result)
````

````python
a = True
b = False
result = a or b and not a
# Сначала отрицание, затем логическое И, затем логическое ИЛИ: True or (False and not True) = True
print(result)
````

Важно помнить о приоритете операций, когда вы комбинируете разные операторы в одном выражении, чтобы избежать нежелательных эффектов и получить ожидаемый результат. В случае сомнения всегда можно использовать скобки, чтобы явно указать порядок выполнения операций.

### Аналог тернарного оператора

Если нам необходим компактный аналог тернарного оператора, можно воспользоваться следующей конструкцией:
```
[значение если условие истина] if [условие] else [значение если условие ложь]
```
Например:

````python
score_1 = 5
score_2 = 0

winner = "Аргентина" if score_1 > score_2  else "Ямайка"

print(winner)
````

Тернарный оператор — это сокращенная форма условного оператора во многих языках программирования. Он полезен для краткой записи простых условных выражений, когда необходимо выбрать одно из двух значений на основе условия. Такой оператор может сделать код более читаемым и компактным.

Однако следует быть осторожными и не злоупотреблять им, чтобы не усложнить код, особенно если условие и оба выражения имеют сложную структуру.

## 3. Оператор match-case

Начиная с версии **3.10** в языке Python появилась такая функциональность как pattern matching (сопоставление шаблонов). Pattern matching представляет собой применение конструкции **match**, которая позволяет сопоставить выражение с некоторым шаблоном. И если выражение соответствует шаблону, то выполняются определенные действия. В этом смысле конструкция match похожа на конструкцию if/else/elif, которая выполняет определенные действия в зависимости от некоторого условия.

Однако функциональность match гораздо шире — она также позволяет извлечь данные из составных типов и применить действия к различным частям объектов.

Конструкция match-case имеет следующее формальное определение:
```
match выражение:
    case шаблон1:
        действие1
    case шаблон2:
        действие2
    ..............
    case шаблонN:
        действиеN
    case _:
        действие_по_умолчанию
```

После ключевого слова match идет сравниваемое выражение. И затем после двоеточия на последующих строках располагаются выражения case. После каждого выражения case указывается шаблон, с которым сравнивается выражение из match. После шаблона через двоеточие указываются набор выполняемых действий блока case.

Конструкция match последовательно сравнивает выражение с шаблонами из блоков case. И если был найден шаблон из какого-нибудь блока case соответствует выражению из match, то выполняются инструкции из данного блока case. В качестве паттернов/шаблонов, с которыми сравниваются выражения, могут применяться как данные примитивных типов, так и последовательности элементов и объектов классов.

Рассмотрим простой пример:

````python
http_error_status = 400
match http_error_status:
    case 400:
        print("Bad request")
    case 404:
        print("Not found")
    case 418:
        print("I'm a teapot")
    case _:
        print("Something's wrong with the Internet")
````

Обратите внимание, что блоки case имеют отступы от начала конструкции match. А инструкции каждого блока case имеют отступы от начала данного блока case. Но если блок case имеет одну инструкцию, ее можно поместить на той же строке, что и оператор case. Также можно определить блок case, который позволяет сравнивать сразу с несколькими знечениями. В этом случае значения разделяются вертикальной чертой:

````python
http_error_status = 500
match http_error_status:
    case 400: print("Bad request")
    case 401 | 403 | 404: print("Not allowed")
    case 418: print("I'm a teapot")
    case _:   print("Something's wrong with the Internet")
````

Кроме того, выражение для сравнения можно обозначить как переменную, и использовать, например, в форматированных строках:

````python
http_error_status = 404
match http_error_status:
    case 400: print("Bad request")
    case 401 | 403 | 404 as status: print(f"Not allowed: {status}")
    case 418: print("I'm a teapot")
    case _:   print("Something's wrong with the Internet")
````

При этом возможности сопоставления с шаблонами у оператора match гораздо шире. Он позволяет при необходимости, например, определять тип объекта в переменной, добавлять дополнительные условия if к шаблонам, или сравнивать элементы последовательностей (списков, кортежей). Дополнительную информацию можно взять [отсюда.](https://docs-python.ru/tutorial/tsikly-upravlenie-vetvleniem-python/konstruktsija-match-case/)

**Итак**:
* Условный оператор if необходим, чтобы некоторый код выполнялся только при определённом условии, которое мы можем задать с помощью логических выражений.
* Оператор match придуман, чтобы делать код более читаемым и понятным в тех случаях, когда нужно проверить некоторое множество возможных значений и выполнить соответствующие действия. Это может быть полезно в работе с перечислениями, при вводе данных и других в случаях, где нужно управлять разветвленной логикой программы.
