Особенности библиотеки Matplotlib:
* Максимально гибкая настройка графиков: изменению подлежат практически все аспекты, включая цвет, стиль линий, маркеры, шрифты и многое другое. Можно использовать различные функции для настройки внешнего вида графика и данных.
* Поддержка разных типов графиков: линейные, точечные, столбчатые графики, круговые диаграммы, гистограммы, и другие. Можно выбрать тип графика, который наилучшим образом подходит для визуализации конкретных данных.
* Поддержка создания субграфиков (subplot) и множественных графиков на одной фигуре позволяет отображать нескольких графиков на одном рисунке для сравнения разных данных или разных метрик над одним набором данных.
* Возможность добавления легенды и аннотации к графикам, для пояснения данных и выделения важных моментов.
* Поддержка создания 3D графиков для визуализации трехмерных данных.
* Создание полноценных интерактивных графиков: поддержка масштабирования, выделения, всплывающих подсказок и автоматического обновления при изменении данных.

Matplotlib поддерживает два метода работы. Один — это процедурный, более простой, и второй — объектно-ориентированный. При процедурном вызывается основная функция модуля pyplot для создания того или иного графика, в которую передаются его параметры. При объектно-ориентированном подходе вы создаете объекты Figure и Axes и затем вызываете методы этих объектов для добавления элементов графика и настройки его параметров. Объектно-ориентированный подход обычно используется для создания более сложных или настраиваемых графиков.  
Поскольку основное свойство библиотеки — упор на простоту использования, будем рассматривать здесь только процедурный метод работы.

**О параметре fmt**

Параметр fmt (от слова "формат") в функции plot() библиотеки Matplotlib используется для определения формата и стиля линий и маркеров на графике. fmt - это строка, которая определяет, каким образом график будет отображаться.

fmt может быть задан следующими пунктами:
* Символ маркера: один символ, который определяет маркеры точек данных (например, 'o' для круглых маркеров, 's' для квадратных и т. д.).
* Стиль линии: один символ, который определяет стиль линии (например, '-' для сплошной линии, '--' для пунктирной, ':' для штриховой и т. д.).
* Цвет: один символ, который определяет цвет линии (например, 'b' для синего цвета, 'r' для красного и т. д.). Допустимые цветовые символы: 'b' (синий), 'g' (зеленый), 'r' (красный), 'c' (голубой), 'm' (пурпурный), 'y' (желтый), 'k' (черный), 'w' (белый).

Комбинируя эти элементы, можно создавать различные стили линий и маркеров для графика. Порядок следования символом при этом значения не имеет. Например:
* `b-` синяя сплошная линия без маркеров.
* `r--` красная пунктирная линия без маркеров.
* `go` зеленые круглые маркеры без линии.
* `cs:` голубые квадратные маркеры со штриховой линией.

Например:

````python
plt.plot(lst, lst_2, alpha=1, color='green', markersize=10, marker='x', linestyle=':', linewidth=1, label='зависимость x от y')
plt.legend()
plt.show()
````

Существует также более детальная настройка каждого из этих параметров.  
Основные параметры функции **plot**:
* x — список или массив значений для координат по оси X.
* y — список или массив значений для координат по оси Y.
* label — подпись для линии, которая будет использоваться в легенде (если легенда включена).
* color — цвет линии (например, 'red' для красной линии).
* linestyle — стиль линии (например, '-' для сплошной линии, '--' для пунктирной, ':' для штриховой и т. д.).
* marker — форма маркера, который будет отображаться на точках данных (например, 'o' для круглых маркеров, 's' для квадратных и т. д.).
* markersize — размер маркеров (если маркеры используются).
* alpha — прозрачность линии (от 0 до 1).
* linewidth или lw — толщина линии.



К диаграмме можно добавить легенду:

### Диаграмма рассеяния (**scatter**)

Точечные графики также можно рисовать с помощью scatter (диаграмма рассеяния). Диаграмма позволяет отобразить данные на графике, где каждая точка представляет пару значений. Точечные графики могут помочь в выявлении паттернов, выбросов, корреляций и других важных характеристик данных, что делает его незаменимым в анализе данных и научной визуализации.

Пример:

````python
# данные первого дня
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,84,88,111,80,103,87,94,78,77,85,86])
plt.scatter(x, y, s=100, marker='x', label='первый день', c='red')

plt.legend()

# данные второго дня
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,87,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, marker='o', s=100, c='green', label='второй день')
plt.legend()

plt.show()
````

Основные параметры функции **scatter**:
* x — список или массив значений для координат по оси X.
* y — список или массив значений для координат по оси Y.
* s — размер маркеров (точек) на графике. Может быть числом или списком значений, чтобы задать размеры разных точек.
* c — цвет маркеров. Может быть строкой (например, 'red'), списком цветов или числовым значением, которое определяет цвет точек в соответствии с цветовой картой (colormap).
* marker — форма маркера, который будет использоваться для каждой точки (например, 'o' для круглых маркеров, 's' для квадратных).
* alpha — прозрачность маркеров (от 0 до 1).
* label — подпись для маркеров, которая будет использоваться в легенде (если легенда включена).

Пример 2.

````python
# Генерируем случайные данные
x_dots = np.random.rand(100)

y_dots = np.random.rand(100)
sizes = np.random.rand(100) * 100  # Размеры маркеров от 0 до 100
colors = np.random.rand(100)  # Цвета маркеров (генерируются случайно)

# Создаем точечный график
plt.scatter(x=x_dots, y=y_dots, s=sizes, c=colors, alpha=0.5, marker='o', label='Случайные данные')

# Добавляем метки осей и заголовок
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Пример точечного графика')

# Включаем легенду
plt.legend()

# Отображаем график
plt.show()
````

Вот несколько примеров использования функции **scatter** в анализе данных:

1. Визуализация распределения данных: с помощью scatter можно отобразить распределение данных на плоскости. Это позволяет быстро определить плотность и разброс данных. Если какие-то точки сгруппированы близко друг с другом, это определенно говорит о наличии некоторых связей между ними.
2. Выявление выбросов: scatter plot позволяет выявлять выбросы или необычные значения в данных. Выбросы часто будут представлены как точки, которые находятся далеко от остальных. Это может быть полезно для идентификации ошибок в данных или необычных событий.
3. Категоризация данных: можно использовать разные маркеры или цвета для точек, чтобы представить разные категории данных. Это помогает визуально сравнивать данные между разными категориями на одном рисунке.

Например, по следующему рисунку мы можем сделать вывод, что данные, нарисованные синим цветом, явно представляют собой отдельную категорию, а две другие группы данных имеют похожие свойства. Кроме того по краям мы видим выбросы данных, которые возможно представляют собой ошибки измерения.

### Столбчатые диаграммы (**bar**)

Функция bar используется для создания столбчатых графиков. Она представляет данные в виде вертикальных (или горизонтальных) столбцов, где каждый столбец представляет значение какой-либо переменной. Столбчатые графики полезны для сравнения данных между различными категориями.

Пример 1.

````python
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([3, 10, 1, 8, 5])

z = np.array(["CD", "BC", "CC", "DC", "EC"])
p = np.array([1, 2, 3, 4, 5])

plt.bar(z, p)
plt.bar(x, y)
plt.show()
````

Основные параметры функции **bar**:
* x — позиции столбцов на оси X (обычно это последовательные числа от 0 до N-1, где N - количество категорий).
* height — высоты столбцов, представляющие данные для каждой категории.
* width — ширина столбцов. Этот параметр можно использовать для настройки ширины столбцов.
* bottom — определяет базовую линию, от которой начинаются столбцы. Это полезно, если вы хотите создать столбцы, начинающиеся не с нуля.
* align — определяет выравнивание столбцов. Параметры могут включать 'center' (по центру), 'edge' (по краям), и т. д.
* color — определяет цвет столбцов.
* label — подпись для столбца, которая будет использоваться в легенде (если легенда включена).
* tick_label — список строк, представляющих метки на оси X для каждой категории.

Пример 2.

````python
# Данные для столбчатой диаграммы
categories = ['A', 'B', 'C', 'D']
values = [10, 30, 15, 25]

# Создаем столбчатую диаграмму
plt.bar(x=categories, height=values, color='#3B7A57', label='Величины',
        tick_label=['Категория A', 'Категория B', 'Категория C', 'Категория D'])

# Добавляем метки осей и заголовок
plt.xlabel('Категории')
plt.ylabel('Значения')
plt.title('Пример столбчатой диаграммы')

# Включаем легенду
plt.legend()

# Отображаем график
plt.show()
````

Вот несколько примеров использования функции **bar** в анализе данных:
1. Сравнение данных: столбчатая диаграмма позволяет сравнивать значения между разными категориями. Например, можно использовать столбчатую диаграмму для сравнения продаж по разным месяцам, рейтингов разных товаров и т.д.
2. Выявление выбросов: путем визуализации данных с помощью столбчатых диаграмм можно быстро выявить выбросы или необычные значения. Если какая-либо категория имеет значительно большее или меньшее значение, чем остальные, это может сигнализировать о проблеме или наоборот об интересном явлении.
3. Сравнение нескольких категорий: можно создавать группы столбцов на одной диаграмме, чтобы сравнить данные между несколькими категориями одновременно.

Например, по следующей диаграмме потребления кофе мы можем понять, у кого из работников предыдущая неделя была гораздо продуктивнее текущей, кто из них всегда много работает, а кто не любит кофе вообще:

### Гистограммы (**hist**)

hist (сокращение от "histogram") - это функция библиотеки Matplotlib, которая используется для создания гистограммы. Гистограмма — это столбчатая диаграмма, которая представляет распределение данных на интервалы (бины) и отображает, сколько значений попадает в каждый интервал.

Пример 1.

````python
# Генерируем случайные данные с нормальным распределением
data = np.random.normal(180, 10, 220)

# Строим гистограмму
plt.hist(data)
plt.show()
````

Основные параметры функции **hist**:
* x — входные данные, для которых вы хотите построить гистограмму.
* bins — этот параметр определяет количество бинов (интервалов) в гистограмме. Можно задать количество бинов явно (например, bins=10), или Matplotlib автоматически разделит данные на бины, если bins не указан.
* range — задает диапазон значений, которые будут участвовать в гистограмме. Например, range=(0, 10) ограничит данные значением от 0 до 10.
* density — если True, гистограмма будет нормализована так, чтобы интеграл ее площади был равен 1. Это полезно, если нужно отобразить относительные частоты вместо абсолютных значений.
* histtype — определяет стиль гистограммы. Возможные значения: 'bar' (столбцы), 'barstacked' (накопленные столбцы), 'step' (лесенка), 'stepfilled' (залитая лесенка).
* color — определяет цвет столбцов гистограммы.
* alpha — устанавливает прозрачность столбцов (от 0 до 1).

Пример 2.

````python
# Генерируем случайные данные
data = np.random.randn(1000)

# Строим гистограмму
plt.hist(x=data, bins=20, range=(-3, 3), density=True, histtype='step', color='blue', alpha=0.5)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Пример гистограммы')
plt.show()
````

Вот несколько примеров использования функции **hist** в анализе данных:
1. Визуализация распределения данных: гистограммы позволяют показывать, как данные распределены вокруг различных значений. Это особенно полезно при исследовании данных и анализе их структуры. Например, гистограмма может показать, как распределены оценки студентов в классе.
2. Выявление выбросов и аномалий. Если у данных есть значения, которые значительно отличаются от остальных, они будут видны на гистограмме как отдельные столбцы.
3. Анализ данных по времени: гистограммы могут быть использованы для анализа данных по времени, например, для определения, когда происходит пиковая активность веб-сайта.
4. *Определение формы распределения: гистограммы могут показать, является ли распределение нормальным, равномерным, экспоненциальным и так далее.

### Круговые диаграммы (**pie**)

pie — это функция библиотеки Matplotlib, которая используется для создания круговых диаграмм (диаграмм-пирогов). Круговая диаграмма представляет собой круг, разделенный на секторы, пропорциональные значениям, которые нужно визуализировать. Эта диаграмма полезна для отображения доли каждой категории в общей сумме данных.

Пример 1.

````python
data = np.array([15, 25, 25, 35, 10])

plt.pie(data)
plt.show()
````

Основные параметры функции **pie**:

* x — входные данные, которые представляют собой список чисел, представляющих доли каждой категории. Например, [10, 20, 30] означает, что первая категория составляет 10% общей суммы, вторая - 20%, и так далее.
* labels — список строк, представляющих подписи для каждой категории. Например, ['Категория A', 'Категория B', 'Категория C'].
* colors — список цветов для каждой категории. Например, ['red', 'green', 'blue'].
* explode — список чисел, определяющих смещение секторов от центра круговой диаграммы. Это позволяет выделить одну или несколько категорий. Например, [0.1, 0, 0] сдвинет первую категорию на 10% от радиуса.
* autopct — строка форматирования, определяющая, как отображать проценты на круговой диаграмме. Например, '%.1f%%' отобразит проценты с одним десятичным знаком.
* shadow — если True, добавляет тень к круговой диаграмме.
* startangle — угол, с которого начинается отрисовка секторов (по умолчанию 0, что означает, что начало секторов будет вверху).

Пример 2.

````python
# Данные для круговой диаграммы
my_sizes = [15, 30, 45, 10]
my_labels = ['Категория A', 'Категория B', 'Категория C', 'Категория D']
my_colors = ['red', 'green', 'blue', 'yellow']
my_explode = (0.1, 0, 0, 0)  # Выделение первой категории из круга

# Строим круговую диаграмму
plt.pie(x=my_sizes, labels=my_labels, colors=my_colors, explode=my_explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Равные пропорции для круга

plt.title('Пример круговой диаграммы')
plt.show()
````

Вот несколько примеров использования функции **pie** в анализе данных:
1. Доли и проценты: круговые диаграммы хорошо подходят для визуализации долей различных частей в отношении целого. Например, можно использовать круговую диаграмму, чтобы показать, какой процент бюджета тратится на разные категории расходов.
2. Сравнение категорий: круговые диаграммы позволяют сравнивать категории данных между собой. Построив груговую диаграмму, можно увидеть, какие категории составляют большую или меньшую часть общего объема.
3. Удобство восприятия: круговые диаграммы обычно легче воспринимаются визуально, чем длинные столбчатые диаграммы, особенно если у вас есть небольшое количество категорий. Круговые диаграммы легко читать и интерпретировать, что делает их доступными для широкого круга пользователей.

#### Другие типы графиков

Естественно, список поддерживаемых в Matplotlib графиков не ограничивается описанным выше. Вот ещё некоторые из наиболее популярных типов графиков, которые можно создать с помощью библиотеки:

1. Трехмерный график (plot3D) и поверхность (surface). Позволяют визуализировать трехмерные данные. Полезны для анализа трехмерных пространственных данных.
2. Диаграмма разброса (scatter3D). Отображает трехмерные точки данных в пространстве. Полезна для визуализации трехмерных данных и поиска корреляций.
3. График рассеяния (scatter) с цветовой картой (cmap). Позволяет отображать точки данных с цветами, зависящими от значения другой переменной. Полезен для визуализации трехмерных данных на двумерной плоскости.
4. Штабированный график (stackplot). Отображает стопки данных друг над другом, чтобы показать, как они суммируются. Полезен для отображения структуры данных.
5. Сетчатка диаграмма (imshow) и тепловая карта (heatmap). Позволяют визуализировать двумерные данные в виде цветовой карты. Полезны для отображения матриц и изображений.
6. Диаграмма ящика и усов (boxplot). Показывает распределение данных в виде ящика с усами. Полезен для выявления выбросов и оценки разброса данных.

#### Сохранение в файл

Для сохранения графика, созданного с помощью библиотеки Matplotlib, в файл, можно использовать метод savefig. Этот метод позволяет сохранить текущую фигуру в различных форматах файлов, таких как PNG, JPEG, PDF, SVG и других. При вызове метода нужно указать путь и имя файла, в который вы хотите сохранить график. Важно также указать желаемый формат файла, добавив расширение файла в качестве части имени. Выбирайте подходящий формат для своей задачи: будет ли это просто рисунок, рисунок для презентации, рисунок для публикации в статье и т.д.

Пример.

````python
# Создание графика
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Сохранение графика в файл (например, в формате PNG)
plt.savefig('my_plot.png')
````

Если используется Google Colab, сохраненный файл можно посмотреть в боковом меню в разделе "Файлы".

Если вы хотите указать конкретный путь для сохранения файла, пропишите полный путь вместе с именем файла:
```
plt.savefig('/полный/путь/к/моему_графику.jpg')
```
При необходимости можно настроить дополнительные параметры при сохранении, такие как разрешение (dpi) или прозрачность фона. Вы можете передать соответствующие аргументы в savefig:
```
plt.savefig('my_plot.png', dpi=300, transparent=True)
```
