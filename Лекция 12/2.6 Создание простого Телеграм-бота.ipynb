## 1. Введение

### Что такое бот?

Бот в Telegram ― это программа, которая взаимодействует с пользователем в мессенджере и выполняет определенные задачи. Telegram-бот умеет делать всё, что мог бы делать человек в чате: отвечать на вопросы, присылать картинки или регистрировать заказ. Главное — уметь его запрограммировать на это.

Само слово «бот» представляет собой сокращение от привычного «робот». Собственно, телеграм-бот как раз им и является. Программа бота общается через специальные Telegram аккаунты, которые создаются автоматически, и дают пользователям возможность выполнять разные задачи посредством чата в мессенджере.

### Что умеют боты?

Telegram боты умееют очень многое. С ними мы можем:
* Развлекаться: получать по своему запросу картинки котиков, анекдоты и мемы.
* Читать новости: бот отправит вам свежие новости, прогноз погоды, курсы валют и другую важную информацию.
* Обмениваться файлами: боты могут вести поиск и присылать различные файлы, например электронные книги.
* Делать покупки: через бота можно разместить заказ и отправить оплату по СБП.
* Бронировать: билеты на транспорт, поездку в такси, билеты в кино, резервацию столиков в ресторане и т.д.
* Быть в курсе событий: бота можно настроить на отправку вам уведомлений и сообщений об ошибках от других программ и сервисов, с которыми вы работаете.
* Работать с нейросетями: переводить текст или генерировать картинки.
* Получать техподдержку: бот может вместо сотрудника отвечать на самые популярные вопросы.
* И многое другое.

При этом боты очевидно обладают некоторыми преимуществами по сравнению с человеком на другом конце чата:
* работают круглосуточно, 7 дней в неделю;
* практически мгновенно отвечают на запрос;
* не нуждаются в установке дополнительного ПО, поскольку любое взаимодействие с ботом осуществляется посредством мессенджера Telegram;
* не затрагивают личные данные без команды пользователя;
* имеют огромное количество функций, ведь его возможности ограниченны порой лишь фантазией и умением программиста.

### Как это работает?

Вот упрощенная схема работы бота:
1. Вы запускаете свою программу с ботом и оставляете работать.
2. Пользователь пишет сообщение боту в Telegram.
3. Сервер Telegram пересылает это сообщение на ваш код (бота).
4. Ваш код на Python обрабатывает сообщение и формирует ответ.
5. Ответ отправляется через Telegram API обратно пользователю.



Для начала разберемся, что такое **API**. API (Application Programming Interface) — это набор правил и инструментов, с помощью которых одна компьютерная программа может общаться с другой через модель **запрос-ответ**.

Можно привести аналогию с меню в кафе:
* Ты (программа 1) видишь список блюд (возможности API).
* Ты делаешь заказ (отправляешь запрос).
* Кухня (программа 2) готовит блюдо и приносит тебе (отправляет ответ).

То есть, API — это интерфейс для общения между программами. Запросы к API пишутся строго определенным образом и позволяют выполнить только то, что предусмотрено в API (т.е. вы не можете попросить приготовить то, чего нет в меню).

Telegram предоставляет специальный API для работы с ботами — Telegram Bot API. Этот API позволяет через запросы:
* Получать сообщения от пользователей.
* Отправлять ответы.
* Отправлять фото, видео, кнопки, стикеры и т.д.
* Управлять ботом (например, настраивать команды).

Эти запросы имеют специальную структуру. Их можно писать вручную, но гораздо проще воспрользоваться специальными библиотеками, которые скрывают сложную структуру и дают готовые шаблоны для запросов (об этом позже).


Теперь разберемся с тем, как сервера Telegram идентифицируют нашего бота.

У обычных пользователей есть уникальный никнейм, по которому человека можно найти в поиске и написать ему. У ботов тоже есть подобный идентификатор, и он называется токен.

Напишем простого бота, который будет получать от поль­зовате­ля тек­сто­вое сооб­щение и воз­вра­щать его обратно (повторять).

После создания файла нужно импортировать библиотеку и ввести токен. Можно сделать это просто сохранив его в строковую переменную:

````python
import telebot

api_token = '6666655555:AAAAAA_BBBBBBBBBB-CCCCCCCCC_DDDDDD'
````

Но очевидно так делать небезопасно, особенно если вы захотите кому-то показать свой код на экране или поделиться файлом.

Поэтому самое время для начала научиться правильно хранить секреты (токены, логины, пароли и т.д.). Один из способов — сохранять такую информацию в специальном файле. Создадим пустой файл со стандартным названием .env (environment, хранилище переменных среды). Откроем его блокнотом и сохраним туда переменную и токен в том же виде, в каком он написан выше. Сейчас, когда у нас есть токен, создадим экземпляр объекта (вспомним уроки по ООП) и сохраним его в переменную:

Теперь взаимодействовать с ботом можно через переменную `bot`.

`message` — это объект из Bot API, содержащий в себе информацию о сообщении. Его некоторые полезные поля это:
* message.chat.id – идентификатор чата,
* message.from.id – идентификатор пользователя,
* message.text – текст сообщения.

Общение пользователя с ботом обычно начинается с команды /start. Научим бота реагировать на эту команду.

Создадим `message handler` – обработчик всех сообщений, которые отправляет пользователь. Обработчик реагирует на входящие сообщения. В нашем примере обработчик будет реагировать на команду /start, поэтому укажем имя функции, которая будет вызывается при наступлении данного события. И напишем саму функцию `start`.

В нашем примере функция `send_message` принимает идентификатор чата (берем его из сообщения) и текст для отправки:
```
@bot.message_handler(commands=['start'])

def start(message):
    bot.send_message(message.chat.id, text=f"Привет, я бот!")
    print(message)
```
В качестве текста для ответа напишем приветственную фразу.

Чтобы получать сообщения от серверов Telegram воспользуемся поллингом (to poll - опрашивать) — постоянным опросом сервера на наличие новых обновлений. Параметр `none_stop` указывает, что мы хотим принимать сообщения постоянно. По умолчанию параметр равен `False`, но мы поставим `True`. Это позволит боту не завершаться и продолжать работу даже в случае возникновения ошибок:

`bot.polling(none_stop=True, interval=0)`

Теперь нужно запустить скрипт целиком и оставить его работать в бесконечном цикле:

````python
# Полный код
bot = telebot.TeleBot(api_token)

@bot.message_handler(commands=['start'])
def start(message):
    bot.send_message(message.chat.id, text="Привет, я бот!")

bot.polling(none_stop=True, interval=0)
````

Теперь добавим функцию echo_all(), которая будет цитировать любое наше сообщение и отвечать на него:

````python
# Версия 1.2
bot = telebot.TeleBot(api_token)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.reply_to(message, "Привет, я бот!")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, message.text)

bot.polling()
````

Переменная **message** является большим словарём, содержащим множество полей. Этот словарь отправляется боту, когда мы пишем ему сообщение. Поле **text** содержит текст отправленного сообщения. Попробуйте заменить message.text на message и перезапустить бота. В качестве сообщения вам вернется весь словарь:

````python
# Версия 1.3
bot = telebot.TeleBot(api_token)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.reply_to(message, "Привет, я бот!")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, message)

bot.polling()
````

Изучите структуру словаря. Можно заметить, что мы отправляем боту все данные о себе, в том числе и свой username, и даже своё имя. Если разобрать структуру, то наше имя доступно через следующие ключи: `{'from_user': {'first_name': 'Evgeniy'}}`.

Давайте поменяем код бота таким образом, чтобы он приветствовал нас по имени:

````python
# Версия 1.4
bot = telebot.TeleBot(api_token)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.send_message(message.chat.id, f"Привет, {message.from_user.first_name}!")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, f'Кажется, вы сказали: "{message.text}"')

bot.polling()
````

### Метод **send_message**

Этот метод отправляет текстовое сообщение в указанный чат. Если chat_id относится к групповому чату или каналу, сообщение будет видно всем участникам группы или подписчикам канала. Метод принимает следующие параметры:

```python
bot.send_message(chat_id, text, parse_mode=None, disable_web_page_preview=False, disable_notification=False, reply_to_message_id=None, reply_markup=None)
```

- `chat_id`: уникальный идентификатор чата, в который будет отправлено сообщение. Это может быть ID пользователя, ID группового чата или ID канала.
- `text`: Текст сообщения, который вы хотите отправить.
- `parse_mode` (необязательный): режим разбора сообщения. Это может быть 'Markdown', 'HTML' или None. Если установлено 'Markdown' или 'HTML', текст сообщения будет интерпретироваться соответствующим образом для форматирования текста. По умолчанию None.
- `disable_web_page_preview` (необязательный): если установлено в True, Telegram не будет создавать предварительный просмотр веб-страниц, включенных в сообщение. По умолчанию False.
- `disable_notification` (необязательный): если установлен в True, бот не будет отправлять уведомление получателю об этом сообщении. По умолчанию False.
- `reply_to_message_id` (необязательный): ID сообщения, на которое вы хотите ответить. Если установлено, сообщение будет отправлено как ответ на указанное сообщение, и Telegram автоматически добавит ссылку на исходное сообщение в целевом чате.
- `reply_markup` (необязательный): объект, представляющий разметку клавиатуры для сообщения. Это может быть InlineKeyboardMarkup, ReplyKeyboardMarkup, ReplyKeyboardRemove или ForceReply.

Пример использования:

```python
bot.send_message(chat_id, "Привет, мир!")
```

Этот пример отправляет сообщение "Привет, мир!" в указанный чат.

### Метод **reply_to**

Данный метод используется для отправки ответа на сообщение, которое было получено от пользователя. Метод принимает следующие параметры:

```python
bot.reply_to(message, text, disable_notification=False, reply_markup=None, parse_mode=None)
```

- `message`: объект Message, который представляет собой сообщение, на которое вы хотите ответить.
- `text`: Текст ответа, который бот отправит пользователю.
- `disable_notification` (необязательный): если установлен в True, бот не будет отправлять уведомление пользователю об этом сообщении. По умолчанию False.
- `reply_markup` (необязательный): объект, представляющий разметку клавиатуры для сообщения. Это может быть InlineKeyboardMarkup, ReplyKeyboardMarkup, ReplyKeyboardRemove или ForceReply.
- `parse_mode` (необязательный): режим разбора сообщения. Это может быть 'Markdown', 'HTML' или None. Если установлено 'Markdown' или 'HTML', текст ответа будет интерпретироваться соответствующим образом для форматирования текста.

Этот метод отправляет сообщение в чат, в котором было получено исходное сообщение, и автоматически цитирует исходное сообщение как часть ответа. Если исходное сообщение было получено в групповом чате, то ответ будет виден всем участникам группы.

Пример использования:

```python
bot.reply_to(message, "Вы написали: " + message.text)
```

Этот пример обрабатывает входящее сообщение и отправляет ответ, включающий текст сообщения, которое было получено от пользователя.

### Команда **/start**

Команда /start в Telegram используется для запуска бота или начальной инициализации его работы. Это одна из стандартных команд, которая часто используется для обеспечения базового взаимодействия между пользователем и ботом. Обычно команду используют для следующих целей:

- Приветствие и инструкции. Когда пользователь впервые начинает использовать вашего бота, командой /start можно поприветствовать пользователя и дать ему инструкцию о том, как использовать бота, в том числе дать список других доступных команд. Правильное приветствие сделает взаимодействие с ботом более дружелюбным и информативным.

- Инициализация настроек. Команда может быть использована для инициализации настроек бота для конкретного пользователя. Например, мы можем сохранить в базе данных информацию о предпочтениях пользователя или его предыдущих взаимодействиях с ботом.

- Отслеживание активности. При получении команды /start мы можем записать факт того, что пользователь начал использовать бота, это поможет отслеживать активность и проводить анализ поведения пользователей.

- Навигация по функциям. В некоторых ботах команда используется как отправная точка для навигации по основным функциям бота. Например, после ввода команды /start бот может предложить пользователю выбрать из нескольких опций или перейти к определенной функции.

### Декоратор **@bot.message_handler**

Декораторы в Python — это механизм, который позволяет изменять поведение функций без изменения их исходного кода.

`@bot.message_handler` — это декоратор в библиотеке telebot, который используется для определения функций-обработчиков сообщений, которые будут вызываться, когда наш бот получает определенный тип сообщения от пользователя.

Этот декоратор привязывает функцию к обработке сообщений определенного типа. Когда бот получает сообщение от пользователя, библиотека telebot автоматически вызывает соответствующую функцию-обработчик, если тип сообщения соответствует указанному в декораторе.

Например, `@bot.message_handler(commands=['start'])` используется для определения функции `send_welcome()`, которая будет вызываться, когда бот получает команду `/start` от пользователя. Точно так же `@bot.message_handler(func=lambda message: True)` определяет функцию `echo_all()`, которая будет вызываться для всех других сообщений, которые получает бот.

Другими словами, мы создаём "виртуальную" функцию message_handler, которая анализирует переданные ей параметры и в зависимости от того, что ей передали, вызыват подходящую "реальную" функцию.

Всё это позволяет боту реагировать на различные типы сообщений, отправляемые пользователями, и предоставляет гибкость в управлении поведением бота в зависимости от содержания сообщений.

Теперь, усвоив справочную информацию, посмотрите еще раз на наш код. Стала ли понятнее каждая строчка в нём?

````python
# Версия 1.4
# Создаем экземпляр класса
bot = telebot.TeleBot(api_token)

# Декоратором оборачиваем функцию, реагирующую на start и help
@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.send_message(message.chat.id, f"Привет, {message.from_user.first_name}!")

# Декоратором оборачиваем функцию, реагирующую на всё кроме команд
@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, f'Кажется, вы сказали: "{message.text}"')

# В бесконечном цикле опрашиваем сервера на наличие новых сообщений
bot.polling()
````
