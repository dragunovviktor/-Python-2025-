## 1. Строки в Python

**Строка** — это тип данных, предназначенный для работы с текстом. Чтобы создать строку в Python, нужно использовать одинарные или двойные кавычки. Для многострочных строк можно использовать тройные кавычки (тоже одинарные или двойные).

```python
first = 'Привет, мир'
second = "Привет, мир"
third = """Привет,
мир"""
```

Строка состоит из последовательности символов. В языке Python нет отдельного символьного типа. Символ — это просто строка длины 1.

Строки могут быть пустыми, то есть не иметь внутри себя ни одного символа. Пробел — это такой же символ, как и все остальные, поэтому следующие две строки не равны:




```python
a = ""   # пустая строка
b = " "  # строка, состоящая из одного пробела
print(a == b)
```

Вводя что-то с командной строки с помощью input(), мы получаем строку. Мы можем убедиться в этом с помощью функции type(). Строка является объектом класса **str**:

```python
a = input("Вводим буквы: ")
print(type(a))
b = input("Вводим числа: ")
print(type(b))
```

Поэтому в предыдущих занятиях приходилось делать явное **преобразование** типов:

```python
a = input()
print(a / 100)
```

```python
a = float(input())
print(a / 100)
```

Можно применять и обратную конвертацию, из числовых переменных int и float в str.

Длину строки можно подсчитать функцией **len**:

```python
my_str = 'My string #154'
len(my_str)
```

В строках можно использовать **экранированные** последовательности:

```python
print("\"Hello,\nworld\"")
```

Весьма удобно для вывода переменных совместно с текстом использовать **форматированную** строку:

```python
name = "Eric"
age = 74
print(f"Hello, {name}. You are {age}.")
```

Ещё одна удобная особенность форматированных строк — возможность вывести значение переменной в виде "имя = значение". Например:

```python
new_counter = 137
print(f"{new_counter = }")
```

Это добавляет "синтаксического сахара" в код и заменяет собой выражение `print("new_counter = ", new_counter)` . Такая конструкция полезна при отладке кода, когда нужно вывести на экран несколько значений переменных вместе с именем, чтобы проверить пошаговую работу алгоритма.

С помощью "r" можно наоборот подавлять экранирование и задавать неформатированную строку (в сыром *raw* виде):

```python
print(r"\"Hello,\nworld\"")
```

### Индексация строк

По индексу можно получить соответствующий ему символ строки. Для этого нужно после самой строки написать в квадратных скобках индекс символа, т.е. работать со строкой как с обычным массивом:

изображение.png

```python
word = 'Привет'
print(word[0])
print(word[5])
```

Если попытаться получить символ по индексу, выходящему за границу строки, Python выдаст ошибку:

```python
print(word[6])
```

В квадратных скобках можно указывать не только фиксированное число, но и считать его с клавиатуры или получить в результате арифметического действия:

```python
word = 'Привет'
print(word[int(input())])
```

В Python разрешены **отрицательные индексы**: word[-1] означает последний символ строки word, word[-2] — предпоследний и т.д.:

```python
print(word[-1])
```

**Внимание!**

Так как в Python строки — это **неизменяемый** тип данных, изменить отдельный символ строки невозможно:

```python
word = 'малоко'
word[1] = 'о'
```

### Перебор элементов строки

Оператор <b>in</b> позволяет проверить наличие подстроки в строке, а также итерировать по строке:

```python
'3.14' in 'Число Пи равно 3.1415926'
```

```python
text = 'Hello'
for c in text:
  print(c)
```

Так как символы в строке пронумерованы, можно перебирать их с помощью **for ... in range()**. Здесь и пригодится функция len():

```python
text = 'Hello'
for i in range(len(text)):
  print(text[i])
```

С помощью конструкции **not in** можно проверять отсутствие подстроки:

```python
txt = "The best things in life are free!"
print("expensive" not in txt)
```

### Юникод. Хранение текстов в памяти компьютера

Поскольку компьютер умеет хранить только двоичные числа, для записи нечисловой информации используются кодовые таблицы — ASCII, Unicode и др.

Изначально в таблице каждому символу был поставлен в соответствие 7-битный код, что позволяло идентифицировать 128 различных символов.

Этого хватало на латинские буквы обоих регистров, знаки препинания и спецсимволы — например, перевод строки или разрыв страницы. Позже код расширили до 1 байта, что позволяло хранить уже 256 различных значений: в таблицу помещались буквы второго алфавита (например, кириллица) и дополнительные графические элементы (псевдографика).

В некоторых относительно низкоуровневых языках (например, в С) можно в любой момент перейти от представления строки в памяти к последовательности байтов, начинающейся по какому-либо адресу.

Сейчас однобайтные кодировки отошли на второй план, уступив место <b>Юникоду</b>.

Юникод содержит соответствия между числом и каким-либо знаком, причем количество знаков может быть любым. Это позволяет одновременно использовать любые символы любых алфавитов и дополнительные графические элементы. Кроме того, в Юникоде каждый символ, помимо кода, имеет некоторые свойства: например, буква это или цифра. Это позволяет более гибко работать с текстами.

В Юникод все время добавляются новые элементы, а сам размер этой таблицы не ограничен и будет только расти, поэтому сейчас при хранении в памяти одного юникод-символа может потребоваться от 1 до 8 байт. Отсутствие ограничений привело к тому, что стали появляться символы на все случаи жизни. Например, такой:

```python
print('\u26FA')
```

В Python существуют встроенные функции ord и chr:
* **ord** — возвращает код символа (от слова *order* — «порядок»).
* **chr** — возвращает по коду соответствующий ему символ.

Посмотрим на десятичный и шестнадцатеричный код нашего эмодзи:

```python
code = ord('⛺')
print("Dec code: ", code)
print("Hex code: ", hex(code))
```

С помощью комбинации этих двух функций мы можем вывести циклом весь алфавит, начиная с буквы "A". Сначала мы получаем десятичный код буквы, добавляем к нему значение счётчика, и переводим обратно в букву чтобы вывести её на экран:

```python
for i in range(26):
    print(chr(ord('A') + i), end=' ')
```

Python также поддерживает Юникод и в названиях переменных:

```python
# не злоупотребляйте подобным
строка = 'my_string'
print(строка)
```

## 2. Срезы строк

Кроме перебора элементов с помощью цикла for ... in range(), взять отдельный её кусок (символы с нечетными номерами и т.д.) можно с помощью срезов (slice).

В самом простом варианте срез строки — ее кусок от одного индекса включительно и до другого не включительно (как для range). То есть это новая, более короткая строка. Срез записывается с помощью квадратных скобок, в которых указывается начальный и конечный индекс, разделенные двоеточием:

```python
text = 'Hello, world!'
print(text[0:5])
print(text[7:12])
```

Если не указан начальный индекс, срез берется от начала (от 0). Если не указан конечный индекс, срез берется до конца строки.

```python
text = 'Hello, world!'
print(text[:5])
print(text[7:])
```

Разрешены отрицательные индексы для отсчета с конца списка.

```python
name = 'Иванов И. И.'
name[:-6]
```

Как и для range, в параметры среза можно добавить третье число — шаг обхода. Этот параметр не является обязательным и записывается через второе двоеточие.

Шаг может быть и <b>отрицательным</b> — для прохода по строке в обратном порядке. Если в этом случае не указать начальный и конечный индекс среза, ими станут последний и первый индексы строки соответственно (а не наоборот, как при положительном шаге):

```python
text = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
text[::-1]
```

### Пример 1.
Из строки '1234567890abcdefghij' выделим все символы до 15-го (не включая).

```python
my_str = '1234567890abcdefghij'
my_str[:15]
```

### Пример 2.
Из строки '1234567890abcdefghij' выделим все символы, начиная с 15-го.

```python
my_str[15:]
```

### Пример 3.
Из строки '1234567890abcdefghij' выделим все символы с четными номерами (нумерация с 0, считаем 0 четным числом). Приходится явно указывать знаки двоеточия:

### Пример 4.
Из строки '1234567890abcdefghij' выделить каждый второй символ, начиная с 3-го (нумерация с 0) и до 17-го (не включая).

## 3. Методы строк

Для эффективной работы со строками в Python реализовано множество методов. Рассмотрим некоторые из них.

### find
Ищет подстроку в строке.

В случае нахождения — возвращает первое вхождение, под которым была найдена подстрока. В случае, если подстрока не найдена, возвращает -1. Значение подстроки чувствительно к регистру. Можно задать область поиска внутри строки, передав вторым и третьим аргументом значения начала среза для поиска и конца среза для поиска.

Если нужно искать от некоторой позиции до конца строки, то указывается только второй аргумент.

Пример ниже ищет первое вхождение строки "is" в строке "This is Python. Python is my favorite language to work with", начиная с пятого символа и до десятого, не включая десятый, и возвращает номер первого символа строки "is" в строке, т.е. 5. Если не задавать область поиска, то получится ответ 2: [0]**T**[1]**h**[2]**i**[3]**s** .

```python
'This is Python. Python is my favorite language to work with'.find('is', 5, 10)
```

Существует и парная функция **rfind**. Работает она так же, но с одним отличием: возвращает последнее вхождение под которым была найдена подстрока.

```python
'This is Python. Python is my favorite language to work with'.rfind('is')
```

**Пример.**

Вводится строка из букв и цифр, программа выводит на экран позиции цифр в строке.

```python
s = input()
for i in range(10):
    print(i, ':', s.find(str(i)))
```

Программа вывела -1 для цифр, которые не встретились во введенной строке. Перепишем, чтобы отсутствующие цифры не выводились:

```python
s = input()
for i in range(10):
    number = s.find(str(i))
    if number != -1:
        print(i, ':', number)
```

Можно заметить, что метод find для цифр 1, 2, 3 нашел только первое вхождение, так как поиск прекращается, как только найден нужный символ.

Рекомендуется использовать find только в том случае, если нужно определить позицию символа. Если требуется только узнать, есть ли такой символ в строке, нужно пользоваться *in*:

```python
s = input()
for i in range(10):
    if str(i) in s:
        print(i, end=' ')
```

### replace
Заменяет одну строку на другую.

Первым параметром — то что заменяем, а вторым — то на что заменяем, исходная строка **не изменяется**.

Если третьим аргументом передано число, то выполняется только такое число замен.

```python
'This is Python. Python is my favorite language to work with'.replace('is', '0')
```

Если хотим только одну замену:

```python
'This is Python. Python is my favorite language to work with'.replace('is', '0', 1)
```

**Примеры.**

Вводится строка из букв и цифр, программа заменяет цифры на точки и выводит на экран результат.

```python
s = "abra123cadabra123"
for i in range(10):
    s = s.replace(str(i), '.')
print(s)
```

Вводится строка из букв и цифр, программа заменяет первое вхождение каждой цифры на соответствующее цифре число точек и выводит на экран результат:

```python
s = "ab4ra3cada2bra1"
for i in range(10):
    s = s.replace(str(i), '.' * i, 1)
print(s)
```

### split
Разбирает строку по указанному разделителю sep (необязательный аргумент метода split, по умолчанию строка разбивается по пробелу) и возвращает результат списком. Можно задать максимальное число разбиений.

```python
sent = "Давайте разобьём предложение на слова"
print(sent.split())
```

**Пример.**

Представить в виде списка строк текст: 'One;two;three;four;five' с разделителем в виде точки с запятой.

```python
'One;two;three;four;five'.split(sep=';')
```

Зададим максимальное число разбиений 2:


```python
'One;two;three;four;five'.split(sep=';', maxsplit=2)
```

Split является одной из самых полезных функций при **парсинге** данных. Она помогает разделить предложения на слова или порезать данные по разделителю, сохранив результат в массив для дальнейшей обработки.

### isdigit
Возвращает True, если все символы строки цифры, причем строка не пустая, иначе возвращает False.

```python
print('12345'.isdigit())
print('1e'.isdigit())
```

### isalpha
Возвращает True, если все символы строки буквы, причем строка не пустая, иначе возвращает False.

```python
print('absd'.isalpha())
print('1e'.isalpha())
```

### upper
Работает с регистром строк, возвращает копию исходной строки в верхнем регистре, исходная строка не изменяется.
### lower
Работает с регистром строк, возвращает копию исходной строки в нижнем регистре, исходная строка не изменяется.

**Пример:** isdigit, isalpha, upper.

По строке 'One1two2three3four4five5' построим две новые строки, первая состоит только из букв строки, записанных в верхнем регистре, а вторая строка состоит только из цифр.

```python
s = 'One1two2three3four4five5'
s = s.upper()
first = ''
second = ''
for i in s:
    if i.isalpha():
        first += i
    elif i.isdigit():
        second += i
print(first, second)
```

### capitalize
Переводит первый символ строки в верхний регистр, а все остальные — в нижний.

```python
s = "пРиВеТ, мИр!"
print(s.capitalize())
```

### count
Подсчитывает количество вхождений одной строки в другую строку. Простейшая форма вызова 'Abracadabra'.count('a') возвращает число вхождений строки 'a' внутри строки 'Abracadabra'.

```python
'Abracadabra'.count('a')
```

При указании трех параметров 'Abracadabra'.count('a', 4, 10) будет выполнен подсчет числа вхождений строки 'a' в срезе [4:10] строки 'Abracadabra'

```python
'Abracadabra'.count('a', 4, 10)
```

### format
Выполняет операцию форматирования строки.

Возвращает копию строки, в которой каждое поле замены подменяется строковым значением соответствующего аргумента.

```python
"The sum of 1 + 2 is {0}".format(1+2)
```
